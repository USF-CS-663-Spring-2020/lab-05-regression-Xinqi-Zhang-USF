```python
import pandas as pd
df = pd.read_csv('credit.csv')
df
```

```python
df.info()
#target: limit
#predictor: Income, Rating, Cards, Age, Education
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### predictor: Income, target: limit

```python
df.plot(kind = "scatter", x = "Income", y="Limit")
```

```python
r = sns.residplot('Limit', 'Income', data=df)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(df[['Income']], df[['Limit']])
```

```python
reg.coef_, reg.intercept_
```

```python
# y = 51.88x + 2389.87
```
